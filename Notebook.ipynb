```python
df = pd.read_csv("Copy of Усі картки здобувачів з ЄДЕБО.xlsx - Аркуш1.csv")
```

```python
groupsData = cleaned_df[["Група","Спеціальність"]].copy()
groupsData = groupsData.dropna()
groupsData["Група"] = groupsData["Група"].apply(lambda x:  str(x)[:2])
groups = alt.Chart(groupsData).mark_bar().encode(
    x= "Група:O",
    y= "count(Група):Q",
    color= "Спеціальність:O",
    tooltip=[alt.Tooltip("Група"),alt.Tooltip("count(Група)",title="кількість"),alt.Tooltip("Спеціальність")]
).transform_filter("datum.Група != null"
                  ).properties(title = "Розподіл по групах")
groups
```

```python
def age(born):
    born = datetime.strptime(born, "%m/%d/%Y").date()
    today = datetime.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
cleaned_df["Вік"] = cleaned_df["Дата народження"].apply(age)
```

```python
genderAgeData = cleaned_df[["Вік","Стать","Спеціальність"]].copy()

Ginterval = alt.selection_point(encodings=['x'])
Finterval = alt.selection_point(encodings=['x'])
ages = alt.Chart(genderAgeData).mark_line().encode(
    x = "Вік",
    y = "count(Вік)",
    tooltip = [alt.Tooltip("Вік"),alt.Tooltip("count(Вік)",title="Кількість")]
).transform_filter(Ginterval).transform_filter(Finterval)
ages1 = alt.Chart(genderAgeData).mark_point().encode(
    x = "Вік",
    y = "count(Вік)",
    tooltip = [alt.Tooltip("Вік"),alt.Tooltip("count(Вік)",title="Кількість")]
).transform_filter(Ginterval).transform_filter(Finterval)


genders = alt.Chart(genderAgeData).mark_bar().encode(
    x = "Стать",
    y = "count(Стать)",
    color = alt.condition(Ginterval, alt.value("blue"), alt.value("lightgrey")),
    tooltip = [alt.Tooltip("Стать"),alt.Tooltip("count(Стать)",title="Кількість")]
).add_params(Ginterval).transform_filter(Finterval)

faculties = alt.Chart(genderAgeData).mark_bar().encode(
    x = "Спеціальність",
    y = "count(Спеціальність)",
    color = alt.condition(Finterval, alt.value("blue"), alt.value("lightgrey")),
    tooltip = [alt.Tooltip("Спеціальність"),alt.Tooltip("count(Спеціальність)",title="Кількість")]
).add_params(Finterval).transform_filter(Ginterval)

ages + ages1 | genders | faculties
```

```python
previous = cleaned_df[["Попередній заклад освіти","Документ про освіту"]].copy()

def finished(certificate):
    if "диплом" in certificate.lower().split():
        return "вища освіта"
    else:
        return "середня освіта"
    
previous["попередня освіта"] = previous["Документ про освіту"].apply(finished)
```

```python
education = alt.Chart(previous).mark_bar().encode(
    x="попередня освіта",
    y="count(попередня освіта)",
    tooltip = [alt.Tooltip("попередня освіта"),alt.Tooltip("count(попередня освіта)",title="Кількість")])
education
```

```python
locations = cleaned_df[["Громадянство","Тип іноземця"]].copy()
locations = locations.loc[locations["Громадянство"]!="Україна"]

nationality = alt.Chart(locations).mark_bar().encode(
    x="Громадянство",
    y="count(Громадянство)",
    color="Тип іноземця",
    tooltip=[alt.Tooltip("Громадянство"),alt.Tooltip("count(Громадянство)",title="Кількість"),alt.Tooltip("Тип іноземця")]
).transform_filter("datum.Громадянство != null")
nationality
```
